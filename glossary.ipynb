```python
from sqlalchemy import Table, create_engine,MetaData, insert,Column,String,Integer,Float,Boolean,select,DateTime,update,and_,or_,not_,distinct
import pandas as pd


engine = create_engine('sqlite:///main.db', echo=False)
metadata = MetaData()
```

```python
filename="english.csv"
#data=pd.read_csv(filename)
data["file"]=filename
data["word_ind"]=data.file+"_"+data.word
data["tags"]=""
all_data=[]
data=data.applymap(str.strip)
for key,rows in data.groupby("word"):
    for i,(x,row) in enumerate(rows.iterrows()):
        rows.loc[x,"word_ind"]+="_"+str(i)
    all_data.append(rows)
data=pd.concat(all_data)

with engine.connect() as con:
    data.to_sql("words",con,if_exists="append",index=False)
```
